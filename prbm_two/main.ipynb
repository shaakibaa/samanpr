```python
#Requirement
import pandas as pd
from statistics import mean
```

```python
data_mt = pd.read_csv("MT_cleaned.csv")
data_vt = pd.read_csv("VT_cleaned.csv")
display(data_mt.head(), data_vt.head())
```

```python
#Question1
data_mt.dropna(subset=['stop_outcome_raw'], inplace=True)
men_driver = len(data_mt[(data_mt['driver_gender']=='M')])
total_driver = len(data_mt['driver_gender'])
print("The proportion of traffic stops in MT involving male drivers:", men_driver/total_driver)
```

```python
#Question2
data_mt.dropna(subset=['stop_outcome_raw', 'stop_outcome'], inplace=True)
arret_outcome = len(data_mt[(data_mt['stop_outcome']=='Arrest')])
arrest_oos = len(data_mt[(data_mt['stop_outcome']=='Arrest')&
                        (data_mt['out_of_state']==True)])
print("Factor increase in a traffic stop arrest likelihood in MT from OOS plates:", 
      (arret_outcome - arrest_oos) / arrest_oos)
```

```python
#Question3
data_mt.dropna(subset=['violation'], inplace=True)
speeding_violation = len(data_mt[data_mt['violation'].str.contains("Speeding")])
total_violation = len(data_mt['violation'])
print("The proportion of traffic stops in MT involving speeding violations:", speeding_violation/total_violation)
```

```python
#Question4
mt_dui = len(data_mt[data_mt['violation'].str.contains("DUI")])
data_vt.dropna(subset=['violation'], inplace=True)
vt_dui = len(data_vt[data_vt['violation'].str.contains("DUI")])
print("Factor increase in traffic stop DUI likelihood in MT over VT:", 
      (mt_dui - vt_dui) / vt_dui)
```

```python
#Question5 (There is no 2020 stop date, therefore I change it to 2010)
vehicle_list = list()
data_mt.dropna(subset=['stop_date', 'vehicle_year'], inplace=True)
vehicle_list = data_mt[data_mt['stop_date'].str.contains("2010")]['vehicle_year'].tolist()
# data_mt['vehicle_year'].tolist()
vehicle_list = [int(i) for i in vehicle_list if i!='NON-' and i!='UNK']
print("The average manufacture year of vehicles stopped in MT in 2010:", int(mean(vehicle_list)))
```
